## Jack's Birthday Hash

We can easily have $p(n) = 1 - \left (1 - \frac{1}{H} \right)^{n}$, then $n(p) = \dfrac{\log(1 - p)}{\log \left(1 - \frac{1}{H} \right)}$

```python
n = 2048
ceil(log(0.5) / log(1 - 1 / n))
```

## Jack's Birthday Confusion

We obtain $p(n) = 1 - \sum_{i = 1}^{n - 1}\left(1 - \dfrac{i}{d} \right)$. 

```python
def same_birthday(n, d):
    p = 1
    for i in range(1, n): #1 -> n-1
        p*=(1-i/d)
    return 1 - p

for i in range(1, n):
    if same_birthday(i, n) >= 0.75:
        print(i)
        break
```
