# Lattices

```python
from sage.all import *
```

## Vectors

```python
v, w, u = vector([2, 6, 3]), vector([1, 0, 0]), vector([7, 7, 2])
3 * (2 * v - w) * (2 * u)
```

## Size and Basis

```python
v = vector([4, 6, 2, 5])
sqrt(v * v)
```

## Gram-Schmidt

```python
from sage.modules.misc import gram_schmidt
v1 = vector([4,1,3,-1])
v2 = vector([2,1,-3,4])
v3 = vector([1,0,-2,7])
v4 = vector([6,2,9,-5])
RR(gram_schmidt([v1, v2, v3, v4])[0][3][1])
```

## What's a Lattice ?

```python
v1 = vector([6, 2, -3])
v2 = vector([5, 1, 4])
v3 = vector([2, 7, 1])
abs(matrix([v1, v2, v3]).det())
```

## Gaussian Reduction

```python
v = vector([846835985, 9834798552])
u = vector([87502093, 123094980])

u, v = matrix([v, u]).LLL()
u * v
```
